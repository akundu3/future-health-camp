### Step 2: Simulating Wearable Patch Data
Real medical sensors experience noise from movement, shivering, or loose patches. Let's write a function to simulate a raw ECG signal combined with a variable heart rate and baseline noise.

```python
def generate_ecg_data(bpm, duration_sec=10, sampling_rate=250, noise_level=0.05):
    """Generates a synthetic ECG signal with adjustable noise and heart rate."""
    total_samples = duration_sec * sampling_rate
    time = np.linspace(0, duration_sec, total_samples)
    
    # Base heart rhythm period
    period = 60.0 / bpm
    
    # Create a synthetic QRS complex (the classic heartbeat spike)
    signal = np.zeros(total_samples)
    for i in range(int(duration_sec * bpm / 60)):
        pulse_center = i * period + 0.2
        # R spike
        signal += 1.2 * np.exp(-((time - pulse_center) / 0.015)**2)
        # Q & S dips
        signal -= 0.25 * np.exp(-((time - (pulse_center - 0.02)) / 0.01)**2)
        signal -= 0.35 * np.exp(-((time - (pulse_center + 0.02)) / 0.01)**2)
        # P and T waves
        signal += 0.15 * np.exp(-((time - (pulse_center - 0.1)) / 0.04)**2)
        signal += 0.25 * np.exp(-((time - (pulse_center + 0.18)) / 0.06)**2)
        
    # Add high frequency random sensor noise
    noise = np.random.normal(0, noise_level, total_samples)
    
    return time, signal + noise

print("🔬 Data Simulation Engine ready.")
```

### Step 3: Interactive UI (The Ultimate Safari Stress Test)
Move the sliders below to simulate different patient heart rates (Beats Per Minute) and background noise from the patch. This utilizes interactive rendering, which is perfect for testing Safari's canvas layout boundaries.

```python
def plot_interactive_ecg(bpm, noise):
    time, ecg_signal = generate_ecg_data(bpm=bpm, noise_level=noise)
    
    plt.figure(figsize=(10, 4))
    plt.plot(time, ecg_signal, color='#d63031', linewidth=1.5, label=f'Wearable Stream ({bpm} BPM)')
    
    # Advanced Engineering Feature: Basic Peak Threshold Detection Line
    threshold = 0.75
    plt.axhline(y=threshold, color='#2d3436', linestyle='--', label='R-Peak Threshold')
    
    plt.title("Live ECG Telemetry - Wearable Patch Simulation", fontsize=14, fontweight='bold')
    plt.xlabel("Time (Seconds)", fontsize=11)
    plt.ylabel("Voltage (mV)", fontsize=11)
    plt.xlim(0, 5) # Show first 5 seconds for visual clarity
    plt.ylim(-0.6, 1.8)
    plt.grid(True, alpha=0.3)
    plt.legend(loc='upper right')
    plt.style.use('ggplot')
    plt.show()
    
    # Simple diagnostic trigger feedback
    if bpm > 140:
        print("🚨 CLINICAL ALERT: Tachycardia Detected! High cardiovascular stress.")
    elif bpm < 50:
        print("⚠️ CLINICAL ALERT: Bradycardia Detected! Low resting rate.")
    else:
        print("💚 Cardiovascular status: Normal rhythm (Safe Zone).")
        
widgets.interact(
    plot_interactive_ecg, 
    bpm=widgets.IntSlider(min=40, max=180, step=5, value=75, description='Patient BPM:'),
    noise=widgets.FloatSlider(min=0.01, max=0.3, step=0.02, value=0.05, description='Patch Noise:')
);
```
